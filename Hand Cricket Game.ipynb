```python
import random

player_choice = input('Odd or Even: ')

first_input = input('Enter a number from 0-6: ')
    
while first_input not in '123456':
    first_input = input("Please enter 0-6: ")

computer_first_input = random.choice('0 1 2 3 4 5 6'.split())

number = int(first_input) + int(computer_first_input)

if number % 2 != 0 and player_choice == 'odd':
    batting_or_bowling = input('Please Choose batting or bowling: ')
elif number % 2 == 0 and player_choice == 'even':
    batting_or_bowling = input('Please Choose batting or bowling: ')
else:
    batting_or_bowling = 'bowling'
    print("You are Bowling")
    
    
game = True
user_score = 0
computer_score = 0

turn = 1


while game:
    user_input = int(input('Enter a number 0-6: '))
    computer_input = random.randint(0,6)
    print('Computer Move= ', computer_input)
    
    if batting_or_bowling == 'batting' and turn == 1:
        if user_input != computer_input:
            user_score += user_input
            print(f"User Score = {user_score} & Computer Score = {computer_score}")
        else:
            print("OUT!!!")
            print(f"Player score {user_score}")
            batting_or_bowling = 'bowling'
            turn = 2
            continue
    elif batting_or_bowling == 'bowling' and turn == 1:
        if user_input != computer_input:
            computer_score += computer_input
            print(f"User Score = {user_score} & Computer Score = {computer_score}")
        else:
            print("OUT!!!")
            print('Computer score= ',computer_score)
            batting_or_bowling = 'batting'
            turn = 2
            continue
            
    if turn == 2 and batting_or_bowling == 'bowling':
        if user_input != computer_input:
            computer_score += computer_input
            print(f"User Score = {user_score} & Computer Score = {computer_score}")
            if computer_score > user_score:
                print('Computer_WON')
                print(f"User Score = {user_score} & Computer Score = {computer_score}")
                game = False
        else:
            print('You Won!!!')
            print(f"User Score = {user_score} & Computer Score = {computer_score}")
            game = False
    elif turn == 2 and batting_or_bowling == 'batting':
        if user_input != computer_input:
            user_score += user_input
            print(f"User Score = {user_score} & Computer Score = {computer_score}")
            if user_score > computer_score:
                print('You Won')
                print(f"User Score = {user_score} & Computer Score = {computer_score}")
                game = False
        else:
            print("Computer Won!!!")
            print(f"User Score = {user_score} & Computer Score = {computer_score}")
    
```
